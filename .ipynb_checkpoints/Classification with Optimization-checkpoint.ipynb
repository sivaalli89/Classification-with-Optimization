This dataset contains information about the operational parameters of various machines and their downtime records

This dataset can be used to analyze machine performance, predict potential failures, and develop predictive maintenance strategies based on operational parameters

### Objective: 

Objective to is to predict the machine failure and prescribe the optimal working conditions to reduce downtime.4%).

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'Machine Downtime.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Removing the null rows
df.dropna(inplace = True)
```

```python
dependent_variables = []

for col in df.columns:
    if df[col].dtype == 'float64':
        dependent_variables.append(col)
    else:
        pass

for col in dependent_variables:
    print(col)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = sns.countplot(df, x = 'Downtime', hue = 'Downtime')
for container in ax.containers:
    ax.bar_label(container)
```

```python
def map(x):
    if x == 'Machine_Failure':
        return 1
    else:
        return 0

df['Downtime'] = df['Downtime'].apply(map)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix

X = df.drop('Downtime', axis = 1)
y = df['Downtime']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
```

```python
def model(X_train, y_train, X_test, y_test, model): 

    preprocess = ColumnTransformer([("Scalar", StandardScaler(), dependent_variables)], remainder = 'drop')   
    pipeline = Pipeline(steps=[('preprocess', preprocess), ('estimator', model)])
    
    model_ = GridSearchCV(estimator = pipeline, param_grid = {}, scoring = 'accuracy', n_jobs = -1, cv = 10)

    model_.fit(X_train, y_train)

    return model_
```

```python
Random_forest = model(X_train, y_train, X_test, y_test, model = RandomForestClassifier())
XGBoost = model(X_train, y_train, X_test, y_test, model = XGBClassifier())
```

```python
test_pred = Random_forest.predict(X_test)
report = classification_report(y_test, test_pred)
matrix = confusion_matrix(y_test, test_pred)

print(report)
print('--'*50)
sns.heatmap(matrix, annot = True,  fmt = '.0f', xticklabels = ['Machine Not Failure', 'Machine Failure'], yticklabels = ['Machine Not Failure', 'Machine Failure'])
plt.title('Random Forest Performance')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
test_pred = XGBoost.predict(X_test)
report = classification_report(y_test, test_pred)
matrix = confusion_matrix(y_test, test_pred)

print(report)
print('--'*50)
sns.heatmap(matrix, annot = True,  fmt = '.0f', xticklabels = ['Machine Not Failure', 'Machine Failure'], yticklabels = ['Machine Not Failure', 'Machine Failure'])
plt.title('XGBoost Performance')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

```python
from pymoo.core.variable import Real

vars = {'Hydraulic_Pressure(bar)': Real(bounds = (df['Hydraulic_Pressure(bar)'].min(), 
                                                  df['Hydraulic_Pressure(bar)'].max())),
 'Coolant_Pressure(bar)':Real(bounds = (df['Coolant_Pressure(bar)'].min(), df['Coolant_Pressure(bar)'].max())),
 'Air_System_Pressure(bar)':Real(bounds = (df['Air_System_Pressure(bar)'].min(), df['Air_System_Pressure(bar)'].max())),
 'Coolant_Temperature':Real(bounds = (df['Coolant_Temperature'].min(), df['Coolant_Temperature'].max())),
 'Hydraulic_Oil_Temperature(?C)':Real(bounds = (df['Hydraulic_Oil_Temperature(?C)'].min(), df['Hydraulic_Oil_Temperature(?C)'].max())),
 'Spindle_Bearing_Temperature(?C)':Real(bounds = (df['Spindle_Bearing_Temperature(?C)'].min(), df['Spindle_Bearing_Temperature(?C)'].max())),
 'Spindle_Vibration(?m)':Real(bounds = (df['Spindle_Vibration(?m)'].min(), df['Spindle_Vibration(?m)'].max())),
 'Tool_Vibration(?m)':Real(bounds = (df['Tool_Vibration(?m)'].min(), df['Tool_Vibration(?m)'].max())),
 'Spindle_Speed(RPM)':Real(bounds = (df['Spindle_Speed(RPM)'].min(), df['Spindle_Speed(RPM)'].max())),
 'Voltage(volts)':Real(bounds = (df['Voltage(volts)'].min(), df['Voltage(volts)'].max())),
 'Torque(Nm)':Real(bounds = (df['Torque(Nm)'].min(), df['Torque(Nm)'].max())),
 'Cutting(kN)':Real(bounds = (df['Cutting(kN)'].min(), df['Cutting(kN)'].max()))  
}
```

```python
from pymoo.core.problem import Problem

class Maintainance(Problem):

    def __init__(self, **kwargs):
        super().__init__(vars = vars, n_obj = 1, **kwargs)

    def _evaluate(self, x, out, *args, **kwargs):

        temp_df = pd.DataFrame(list(x))

        obj = XGBoost.predict_proba(temp_df)
        obj = [a[1] for a in obj]

        out['F'] = np.column_stack([obj])
```

```python
from pymoo.core.mixed import MixedVariableMating, MixedVariableSampling, MixedVariableDuplicateElimination
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize

problem = Maintainance()

algorithm = NSGA2(pop_size = 100,
                 sampling = MixedVariableSampling(),
                 mating = MixedVariableMating(eliminate_duplicates = MixedVariableDuplicateElimination()),
                 eliminate_duplicates = False)

res = minimize(problem,
              algorithm,
              ('n_gen', 100), verbose = False, seed = 42)
```

```python
results = pd.DataFrame(list(res.X))
results['Maintainence'] = res.F
```
